```python
import ast
import pandas as pd
final = {}
title = ["Message Player 0: Fire", "Message Player 0: No Fire", "Message Player 1: Fire", "Message Player 1: No Fire", "Message Player 2: Fire", "Message Player 2: No Fire",
 "Agent Role (1 Defender)", "History", "Board: player 0 empty", "Board: player 0 fire", "Board: player 0 gold", 
 "Board: player 1 empty", "Board: player 1 fire", "Board: player 1 gold", 
  "Board: player 2 empty", "Board: player 2 fire", "Board: player 2 gold",
  "Score empty",  "Score fire", "Score gold",  "Hand Active Player empty",  "Hand Active Player fire", "Hand Active Player gold", "Active Player",
   "Number", "Regret Sum", "Strategy_Sum", "Strategy", "Action 1", "Action 2" ]
for n, i in enumerate(k.nodes):
    state = ast.literal_eval(i)
    info = [k.nodes[i].number,k.nodes[i].regret_sum,k.nodes[i].strategy_sum,k.nodes[i].strategy, k.nodes[i].get_average_strategy()[0],k.nodes[i].get_average_strategy()[1]]
    print(len(state))
    if len(state)==18:
        state = ["0"]*6 + state
    elif len(state)==20:
        state = state[:2]+["0"]*4 + state[2:]
    elif len(state)==22:
        state = state[:4]+["0"]*2 + state[4:]
    
    print(state)
    final[n] = state+info
    print(len(state)) 
       
df_results = pd.DataFrame(final).T
df_results.columns = title
df_results.to_excel("strategy.xlsx")
```
